#  Matplotlib: Line Plot

## What is a Line Plot?
- A **line plot** displays data points connected by straight lines.
- Useful for showing:
  - **Trends** over time
  - **Relationships** between variables
  - **Continuous data**

**Common Options:**
- color="blue" or short code "r", "g", "b"
- linestyle="--" ("-", ":", "-. " also possible)
- marker="o" ("s", "x", "^" etc.)
- linewidth=2
- markersize=8


````python
month=['Jan', 'Feb', 'mar', 'Apr','May','Jun']
recovery_rate = [65, 70, 60, 78, 80, 85]

plt.plot(month, recovery_rate, marker='o', linestyle='-', color='b', label="Recovery Rate")

plt.xlabel('Month')
plt.ylabel('Recovery Percentage')
plt.title("Patien Recovery Trends")
plt.legend()
plt.show()
````

# Bar Plot (Matplotlib)

---

## 1. What is a Bar Plot?
- A **bar plot (or bar chart)** is used to represent **categorical data** with rectangular bars.  
- The **length/height** of the bar corresponds to the value of the category.  

---

## 2. Key Points
- `plt.bar(x, height)` → Creates a **vertical bar plot**.  
- `plt.barh(y, width)` → Creates a **horizontal bar plot**.  
- **Categories** (x-axis or y-axis) → represent labels.  
- **Values** (y-axis or x-axis) → represent numeric data.  
- Useful for comparing **different groups/categories**.  

---

## 3. Important Parameters in `plt.bar()`
- **`x`** → categories or positions of bars.  
- **`height`** → values for each bar.  
- **`color`** → set bar colors.  
- **`width`** → thickness of bars (default = 0.8).  
- **`label`** → name for legend.  
- **`edgecolor`** → border color of bars.  
- **`align`** → bar alignment (`'center'` or `'edge'`).  

---

## 4. Bar Plot Types
- **Vertical Bar Plot** → `plt.bar()`  
- **Horizontal Bar Plot** → `plt.barh()`  
- **Grouped Bar Plot** → compare multiple datasets side by side.  
- **Stacked Bar Plot** → bars stacked on top of each other.  

---


## Python Implementation

````python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 40]

# Simple Vertical Bar Plot
plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
````

## Histogram Bar Plot

````python
plt.barh(categories, values, color='orange')
plt.title("Horizontal Bar Plot")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.show()
````

## Grouped Bar Plot

````python
import numpy as np

categories = ['A', 'B', 'C', 'D']
values1 = [10, 24, 36, 40]
values2 = [20, 18, 30, 25]

x = np.arange(len(categories))  # positions
width = 0.4

plt.bar(x - width/2, values1, width, label='Set 1', color='lightgreen')
plt.bar(x + width/2, values2, width, label='Set 2', color='salmon')

plt.xticks(x, categories)
plt.title("Grouped Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.legend()
plt.show()
````

## Stacked bar Plot

````python
categories = ['A', 'B', 'C', 'D']
values1 = [10, 24, 36, 40]
values2 = [20, 18, 30, 25]

plt.bar(categories, values1, label='Set 1', color='lightblue')
plt.bar(categories, values2, bottom=values1, label='Set 2', color='pink')

plt.title("Stacked Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.legend()
plt.show()
````

# Pie Chart (Matplotlib)

---

## 1. What is a Pie Chart?
- A **pie chart** is a circular statistical graphic divided into slices.  
- Each slice represents a **proportion/percentage** of the whole.  
- Useful for showing **parts of a whole** (composition).  

---

## 2. Key Points
- `plt.pie(data)` → Creates a pie chart.  
- Values are automatically converted into **proportions** of the total.  
- By default, slices are drawn **counterclockwise**.  

---

## 3. Important Parameters in `plt.pie()`
- **`x`** → Data values (list/array).  
- **`labels`** → Names for each slice.  
- **`colors`** → Set custom slice colors.  
- **`autopct`** → Display percentage (`'%1.1f%%'` for 1 decimal place).  
- **`startangle`** → Rotate the start of the pie chart (e.g., `90` starts from top).  
- **`explode`** → Offset a slice for emphasis (list of values).  
- **`shadow`** → Add shadow effect.  
- **`counterclock`** → Plot clockwise or counterclockwise.  

---

## 4. Pie Chart Variants
- **Simple Pie Chart** → Basic representation.  
- **Exploded Pie Chart** → Highlight a slice by pulling it outward.  
- **Percentage Pie Chart** → Show percentage on slices with `autopct`.  
- **Donut Chart** → Pie chart with a hole in the middle (using `wedgeprops`).  

---


## Percentage Pie Chart

````python
sizes = [40, 30, 20, 10]
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Percentage Pie Chart")
plt.show()
````

## Exploded Pie Chart

````python
sizes = [50, 25, 15, 10]
labels = ['Python', 'C++', 'Java', 'Others']
explode = [0.1, 0, 0, 0]  # "explode" 1st slice

plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140)
plt.title("Exploded Pie Chart")
plt.show()
````

## Donut pie Chart

````python
sizes = [35, 25, 25, 15]
labels = ['Red', 'Blue', 'Green', 'Yellow']

# Create pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops={'width':0.4})
plt.title("Donut Chart")
plt.show()
````

# Bar Plot vs Pie Chart

| Feature              | Bar Plot                                                | Pie Chart                                              |
|----------------------|---------------------------------------------------------|-------------------------------------------------------|
| **Purpose**          | Compare values across different categories.             | Show proportions/percentages of a whole.              |
| **Data Type**        | Categorical (discrete values).                          | Categorical (part-to-whole relationships).            |
| **Best For**         | Comparing sizes of categories side by side.             | Showing contribution of each category to the total.   |
| **Readability**      | Easy to read, even with many categories.                | Becomes confusing with too many slices (>6).          |
| **Representation**   | Uses rectangular bars (length/height = value).          | Uses circular slices (angle/area = value).            |
| **Variants**         | Vertical, Horizontal, Grouped, Stacked.                 | Simple, Percentage, Exploded, Donut.                  |
| **Accuracy**         | More accurate for comparing differences between values. | Less accurate, harder to compare close values.        |
| **Example Use Case** | Comparing sales of products across months.              | Showing market share of different companies.          |

# Scatter Plot (Matplotlib)

---

## 1. What is a Scatter Plot?
- A **scatter plot** is used to represent the relationship between **two continuous variables**.  
- Each point on the graph represents a **data observation (x, y)**.  
- Useful for **correlation analysis, outlier detection, trend visualization**.  

---

## 2. Key Points
- `plt.scatter(x, y)` → Creates a scatter plot.  
- Shows how one variable is affected by another.  
- Often used in **exploratory data analysis (EDA)** in Data Science.  

---

## 3. Important Parameters in `plt.scatter()`
- **`x, y`** → Data points (arrays/lists).  
- **`c`** → Color of points (single value or colormap).  
- **`s`** → Size of points.  
- **`alpha`** → Transparency of points (0 to 1).  
- **`marker`** → Shape of points (e.g., `o`, `^`, `s`).  
- **`cmap`** → Colormap when `c` is an array.  
- **`edgecolors`** → Color of marker edges.  

---

## 4. When to Use Scatter Plot?
- To **visualize correlation** (positive, negative, or no correlation).  
- To **detect outliers** in data.  
- To see **distribution patterns** (clusters, spread, etc.).  
- Common in **Machine Learning & Data Science** for feature relationships.  

---

## 5. Python Examples

### 🔹 Simple Scatter Plot
```python
import matplotlib.pyplot as plt

# Data
x = [5, 7, 8, 7, 6, 9, 5, 6]
y = [99, 86, 87, 88, 100, 86, 103, 87]

plt.scatter(x, y, color='blue')
plt.title("Simple Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()


## Scatter Plot with Custom Size & Color

````python
x = [5, 7, 8, 7, 6, 9, 5, 6]
y = [99, 86, 87, 88, 100, 86, 103, 87]
sizes = [20, 50, 100, 200, 250, 300, 400, 450]  # point sizes

plt.scatter(x, y, s=sizes, color='green', alpha=0.5, edgecolors='black')
plt.title("Scatter Plot with Sizes")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
````

## Scatter Plot with Colormap

````python
import numpy as np

# Random data
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)   # color values
sizes = 500 * np.random.rand(50)  # point sizes

plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.colorbar()  # show color scale
plt.title("Scatter Plot with Colormap")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
````

## Multiple Scatter Plots (Comparing Groups)

````python
# Group 1
x1 = [5, 7, 8, 7]
y1 = [99, 86, 87, 88]

# Group 2
x2 = [6, 9, 5, 6]
y2 = [100, 86, 103, 87]

plt.scatter(x1, y1, color='red', label="Group 1")
plt.scatter(x2, y2, color='blue', label="Group 2")
plt.title("Multiple Scatter Plots")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
````

# Histogram (Matplotlib)

---

## 1. What is a Histogram?
- A **histogram** is used to represent the **distribution of continuous data**.  
- Data is divided into intervals called **bins**, and the frequency of values in each bin is shown as a bar.  
- Unlike a bar plot (for categorical data), a histogram is for **numerical data distribution**.  

---

## 2. Key Points
- `plt.hist(data)` → Creates a histogram.  
- X-axis shows **bins (intervals)**, Y-axis shows **frequency (count of values)**.  
- Useful for **understanding data distribution**, **detecting skewness**, and **outliers**.  

---

## 3. Important Parameters in `plt.hist()`
- **`x`** → Input data (list/array).  
- **`bins`** → Number of bins or sequence of bin edges.  
- **`color`** → Fill color of bars.  
- **`edgecolor`** → Border color of bars.  
- **`alpha`** → Transparency of bars.  
- **`density=True`** → Normalize histogram (area = 1, shows probability density).  
- **`histtype`** → Style (`'bar'`, `'step'`, `'stepfilled'`).  

---

## 4. When to Use Histogram?
- To **visualize data distribution**.  
- To check if data is **normally distributed**.  
- To **detect skewness** (left/right skew).  
- To **find outliers** in datasets.  
- Widely used in **statistics & data science**.  

---

## 5. Python Examples

### Simple Histogram
```python
import matplotlib.pyplot as plt

data = [1, 1, 2, 2, 2, 3, 3, 4, 4, 5, 5, 5, 5, 6, 7, 8, 9, 9, 10]

plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title("Simple Histogram")
plt.xlabel("Bins (Intervals)")
plt.ylabel("Frequency")
plt.show()


## Histogram with More Bins

````python
import numpy as np

# Random normal distribution
data = np.random.randn(1000)

plt.hist(data, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title("Histogram with 30 Bins")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
````

## Normalized Histogram (Probability Density)

````python
plt.hist(data, bins=20, color='salmon', density=True, edgecolor='black', alpha=0.6)
plt.title("Normalized Histogram (Probability Density)")
plt.xlabel("Value Range")
plt.ylabel("Probability Density")
plt.show()
````

## Multiple Histograms (Comparing Distributions)

````python
data1 = np.random.normal(0, 1, 1000)  # mean=0, std=1
data2 = np.random.normal(2, 1.5, 1000)  # mean=2, std=1.5

plt.hist(data1, bins=30, alpha=0.5, label="Dataset 1")
plt.hist(data2, bins=30, alpha=0.5, label="Dataset 2")
plt.title("Comparing Two Histograms")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.legend()
plt.show()
````

# on Word Cloud (Python)

---

## 1. What is a Word Cloud?
- A **Word Cloud** (or Tag Cloud) is a **visual representation of text data**.  
- Words are displayed with **size proportional to their frequency**.  
- Useful in **text analysis, NLP, and data visualization** to identify **important/most frequent words**.  

---

## 2. Key Points
- Requires **`wordcloud`** library in Python.  
- Works well with **large text datasets**.  
- Can be customized by **shape, color, background, font**.  
- Often used in **EDA of textual data**.  

---

## 3. Important Parameters in `WordCloud()`
- **`width`** → Width of the image.  
- **`height`** → Height of the image.  
- **`background_color`** → Background color of the word cloud.  
- **`max_words`** → Maximum number of words to display.  
- **`stopwords`** → List of words to ignore (like "the", "and").  
- **`colormap`** → Color map for words (e.g., `'viridis'`, `'plasma'`).  
- **`mask`** → Shape of the word cloud (use custom image).  

---

## 4. When to Use Word Cloud?
- To **visualize the most frequent words** in text data.  
- To **quickly explore text data** for insights.  
- Common in **social media analysis, reviews, articles, NLP projects**.  

---




# Python Examples with Inferences

###  Simple Word Cloud

````python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "Data Science is amazing. Python is amazing. Machine Learning is amazing."

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Simple Word Cloud")
plt.show()
````

Inference:

The word “amazing” appears largest → most frequent.

Quickly shows which words are dominant in the text.

Useful for getting a quick visual summary of key terms.

## Word Cloud with Stopwords

````python
from wordcloud import STOPWORDS

text = "Python is great. Python is easy. Python is powerful. Data Science with Python."

stopwords = set(STOPWORDS)
stopwords.update(["is", "with"])  # add custom stopwords

wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords).generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud with Stopwords Removed")
plt.show()
````

Inference:

Stopwords like “is” and “with” are ignored → focus on meaningful words.

“Python” is clearly the dominant word.

Helps filter out noise in textual data.

## Word Cloud with Colormap

````python
text = "Data Visualization is key in Data Science. Word Cloud helps in visualization."

wordcloud = WordCloud(width=800, height=400, background_color='black', colormap='plasma').generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud with Colormap")
plt.show()
````

Inference:

Colormap enhances visual appeal and highlights words better.

Important words like “Data”, “Visualization” stand out.

Useful when presenting insights in reports or presentations.
